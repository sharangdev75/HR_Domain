```python
encoded_var = pd.get_dummies(hr_df_main[hr_df1], drop_first = True)
```

```python
encoded_var.drop(['SLNO', 'Candidate Ref'], axis = 1, inplace = True)
```

```python
 hr_df_main['Status'].replace(['Joined', 'Not Joined'],[1,0], inplace = True)
```

```python
hr_df_main.head(20)
```

```python
Y = hr_df_main.Status
X = sm.add_constant(encoded_var)
```

```python
train_X,test_X,train_Y,test_Y = train_test_split(X,Y, test_size = 0.2, random_state = 40)
```

```python
logit = sm.Logit(train_Y, train_X)
```

```python
logit_model = logit.fit()
```

```python
logit_model.summary2()
```

```python
def p_value(lm):
    var_values = pd.DataFrame(lm.pvalues)
    var_values['vars'] = var_values.index
    var_values.columns = ['pvals', 'vars']
    return list(var_values[var_values.pvals<=0.05]['vars'])
    
```

```python
signi_vars = p_value(logit_model)
```

```python
signi_vars
```

```python
logit_model_vars = sm.Logit(train_Y, sm.add_constant(train_X[signi_vars])).fit()
```

```python
logit_model_vars.summary2()
```

```python
y_pred_df = pd.DataFrame({"Actual": test_Y, "prediction": logit_model_vars.predict(sm.add_constant(test_X[signi_vars]))})
```

```python
y_pred_df.sample(10, random_state = 40)
```

```python
y_pred_df['predict'] = y_pred_df.prediction.map(lambda x: 1 if x > 0.5 else 0)
y_pred_df.sample(10, random_state = 50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
from sklearn import metrics
```

```python
def cm(actual, predict):
    cm = metrics.confusion_matrix(actual, predict, [1,0])
    sn.heatmap(cm, annot = True, fmt = ' .2f', xticklabels = ['Bad credit', 'Good Credit'], yticklabels = ['Bad Credit', 'Good Credit'])
    plt.ylabel('True')
    plt.xlabel('predicted')
    plt.show
```

```python
cm(y_pred_df.Actual, y_pred_df.predict)
```

```python
print(metrics.classification_report(y_pred_df.Actual, y_pred_df.predict))
```
